## Load CSV Data

### Subtask:
Load the CSV file from the specified Google Drive path into a pandas DataFrame.


**Reasoning**:
I need to import the pandas library, define the CSV file path, load the CSV into a pandas DataFrame, and then display the first few rows of the DataFrame to complete the subtask.



```python
import pandas as pd

csv_file_path = '/content/drive/MyDrive/2020_Neighborhood_Tabulation_Areas_(NTAs)_20260116.csv'
df = pd.read_csv(csv_file_path)

print("CSV file loaded into DataFrame. First 5 rows:")
df.head()
```

**Reasoning**:
The 'the_geom' column contains geometric data in a string format. To convert this DataFrame into a Shapefile, I first need to parse this string representation into proper geospatial objects using a library like `shapely` and then create a GeoDataFrame.



```python
import geopandas as gpd
from shapely import wkt

# Convert the 'the_geom' column from string to WKT (Well-Known Text) geometric objects
df['geometry'] = df['the_geom'].apply(wkt.loads)

print("Converted 'the_geom' column to 'geometry' with WKT objects. First 5 geometries:")
print(df['geometry'].head())
```

**Reasoning**:
Now that the geometric data is parsed into `shapely` objects in the 'geometry' column, I need to convert the pandas DataFrame `df` into a GeoDataFrame, specifying the 'geometry' column and its Coordinate Reference System (CRS) for proper geospatial representation.



```python
geodf = gpd.GeoDataFrame(df, geometry='geometry', crs='EPSG:4326')

print("Created GeoDataFrame. First 5 rows with geometry:")
geodf.head()
```

**Reasoning**:
The GeoDataFrame `geodf` has been created with the correct geometry and CRS. The next step is to save this GeoDataFrame as a Shapefile, which is the final output required by the main task.



```python
output_shapefile_path = '/content/2020_Neighborhood_Tabulation_Areas.shp'
geodf.to_file(output_shapefile_path)

print(f"Shapefile successfully created at: {output_shapefile_path}")
```

**Reasoning**:
The previous step encountered warnings about column name truncation and geometry string truncation when saving to a Shapefile. This is due to limitations of the ESRI Shapefile format. To resolve these, I need to rename columns to be 10 characters or less and drop the original 'the_geom' column since its parsed geometry is already present in the 'geometry' column.



```python
geodf_cleaned = geodf.drop(columns=['the_geom']).copy()
geodf_cleaned = geodf_cleaned.rename(columns={
    'BoroCode': 'BoroCode',
    'BoroName': 'BoroName',
    'CountyFIPS': 'CntyFIPS',
    'NTA2020': 'NTA_2020',
    'NTAName': 'NTA_Name',
    'NTAAbbrev': 'NTA_Abbrev',
    'NTAType': 'NTA_Type',
    'CDTA2020': 'CDTA_2020',
    'CDTAName': 'CDTA_Name',
    'Shape_Length': 'Shape_Len',
    'Shape_Area': 'Shape_Area'
})

print("GeoDataFrame cleaned for Shapefile compatibility.")
print(geodf_cleaned.columns)
```

**Reasoning**:
The GeoDataFrame `geodf_cleaned` is now properly formatted with compatible column names and geometric objects. The final step is to save this cleaned GeoDataFrame to a Shapefile to complete the task.



```python
output_shapefile_path = '/content/2020_Neighborhood_Tabulation_Areas_cleaned.shp'
geodf_cleaned.to_file(output_shapefile_path)

print(f"Cleaned Shapefile successfully created at: {output_shapefile_path}")
```
