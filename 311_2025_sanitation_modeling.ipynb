## **2. 311_Seasonal_Analysis**

```python
trash_complaint_types = {
    "Dirty Condition",
    "Missed Collection",
    "Litter Basket Complaint",
    "Litter Basket Request",
    "Recycling Basket Complaint",
    "Residential Disposal Complaint",
    "Commercial Disposal Complaint",
    "Institution Disposal Complaint",
    "Dumpster Complaint",
    "Illegal Dumping",
    "Street Sweeping Complaint",
    "Sanitation Worker or Vehicle Complaint",
#    "Transfer Station Complaint",
#    "DSNY Internal",
}

sewer_complaint_types = {
    "Sewer",
    "Root/Sewer/Sidewalk Condition",
    "Indoor Sewage",
#    "Industrial Waste",
#    "WATER LEAK",
#    "Water System",
}

print("Trash Complaint Types:", trash_complaint_types)
print("Sewer Complaint Types:", sewer_complaint_types)
```

```python
import numpy as np

problem_col = "Problem (formerly Complaint Type)"

def trash_bucket(problem):
    if problem in {"Dirty Condition", "Street Sweeping Complaint"}:
        return "Street Cleanliness"
    elif problem in {"Missed Collection", "Sanitation Worker or Vehicle Complaint"}:
        return "Collection Service Gaps"
    elif problem in {
        "Illegal Dumping", "Residential Disposal Complaint",
        "Dumpster Complaint", "Commercial Disposal Complaint",
        "Institution Disposal Complaint"
    }:
        return "Illegal Dumping & Bulk Waste"
    elif problem in {"Litter Basket Request", "Litter Basket Complaint", "Recycling Basket Complaint"}:
        return "Bin Capacity"
    else:
        # Default for complaints not explicitly categorized into the new trash buckets
        return "other_unclassified"

# Ensure 'Created Date' is in datetime format and 'Month' is extracted
df_sanitation['Created Date'] = pd.to_datetime(df_sanitation['Created Date'])
df_sanitation['Month'] = df_sanitation['Created Date'].dt.month

# Apply trash_bucket for initial categorization to all of df_sanitation
df_sanitation["Sub_Complaints_Type"] = df_sanitation[problem_col].apply(trash_bucket)

# Synchronize sewer and flooding complaints with their original Problem types
sewer_mask = df_sanitation[problem_col].isin(sewer_complaint_types)
df_sanitation.loc[sewer_mask, 'Sub_Complaints_Type'] = df_sanitation.loc[sewer_mask, problem_col]

# Re-define flooding_complaint_types as it was commented out in a previous cell update.
# Based on the last executed context, it should be {'Standing Water'}.
flooding_complaint_types = {'Standing Water'}
flooding_mask = df_sanitation[problem_col].isin(flooding_complaint_types)
df_sanitation.loc[flooding_mask, 'Sub_Complaints_Type'] = df_sanitation.loc[flooding_mask, problem_col]

# Now, filter the specific DataFrames from the updated df_sanitation
df_311_trash = df_sanitation[df_sanitation[problem_col].isin(trash_complaint_types)].copy()
df_311_sewer = df_sanitation[df_sanitation[problem_col].isin(sewer_complaint_types)].copy()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_monthly_distribution(df, title, color):
    if not df.empty:
        monthly_counts = df['Month'].value_counts().sort_index()
        plt.figure(figsize=(10, 6))
        sns.barplot(x=monthly_counts.index, y=monthly_counts.values, color=color, legend=False)
        plt.title(title)
        plt.xlabel('Month')
        plt.ylabel('Number of Complaints')
        plt.xticks(ticks=range(0, 12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
    else:
        print(f"DataFrame for '{title}' is empty, cannot generate plot.")

# Generate monthly bar plot for df_311_trash
plot_monthly_distribution(df_311_trash, 'Monthly Distribution of 311 Trash Complaints', 'forestgreen')

# Generate monthly bar plot for df_311_sewer
plot_monthly_distribution(df_311_sewer, 'Monthly Distribution of 311 Sewer Complaints', 'steelblue')
```
